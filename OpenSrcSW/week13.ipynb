```python
import csv
import matplotlib.pyplot as plt
f=open('sseoul.csv')
data=csv.reader(f)
next(data)
Aug=[]
Jan=[]
for row in data:
    month = row[0].split('-')[1]
    if row[-1]!='':
        if month=='08':
            Aug.append(float(row[-1]))
        elif month=='01':
            Jan.append(float(row[-1]))
plt.hist(Aug,bins=100,color='red',label='Aug')
plt.hist(Jan,bins=100,color='blue',label='Jan')
plt.legend()
plt.show()
```
